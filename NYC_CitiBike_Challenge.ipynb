```python
# import required modules
import pandas as pd
import numpy as np
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
filepath = r'C:\Users\billy\CitiBike_Challenge'
in_file = r'\201908-citibike-tripdata.csv'
out_file = r'\201908-citibike-tripdata_dt.csv'
summary_file = r'\bike_summary.csv'

full_in_file = f'{filepath}{in_file}'
full_out_file = f'{filepath}{out_file}'
full_summary_file = f'{filepath}{summary_file}'

df_trip = pd.read_csv(full_in_file)
df_trip.head()
```

```python
# 2. Check the datatypes of your columns. 
df_trip.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
df_trip['tripduration_dt'] = pd.to_datetime(df_trip['tripduration'],
                                            unit='m')
```

```python
# Additional column added to be able to filter by number of days instead of just hours
df_trip['totaldays_tripduration'] = (df_trip['tripduration'] / 1440).apply(np.floor).astype(int)
```

```python
# Preview trip data df
df_trip
```

```python
# 4. Check the datatypes of your columns. 
df_trip.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
df_trip.to_csv(full_out_file,index=False)
```

```python
# Create summary df for review against Tableau output
df_summary = df_trip.groupby('tripduration').count()[['starttime']].reset_index().rename(columns={'starttime': 'Count'})
```
